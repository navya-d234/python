```python
# Question 11: Find the most frequent character in "mississippi"
str=input("enter a string")
max_char=''
ma_count=0
for char in set(str):
    count=str.count(char)
    if count>ma_count:
        ma_count=count
        max_char=char
print(f"The most frequent character in {str} is {max_char} the number of time is repeated is {ma_count}")
```

```python
# Question 12: Check if two strings are anagrams: "listen" and "silent"
str1=input("enter string one")
str2=input("enter String two")
if sorted(str1)==sorted(str2):
    print("the two string is anagram")
else:
    print("Not a anagram")
```

```python
# Question 14: Count consonants in "Hello World"
str=input("enter string")
count=0
vowels='aeiouAEIOU'
for s in str:
    if s.isalpha() and s.lower() not in vowels:
        count+=1
print(f"The number consonants in the String is {count}")
```

```python
# Question 15: Find the longest word in "Python is a programming language"
str=input("enter sentence")
words=str.split()
longest=''
for word in words:
    if len(word)>len(longest):
        longest=word
print(f"The longest word in the sentence is {longest}")
```

```python
# Question 16: Remove all punctuation from "Hello, World! How are you?"
import string
str=input("enter a sentence")
cleaned_sen=''.join(ch for ch in str if ch not in string.punctuation)
print(f"The removed puntuation in the sentence is {cleaned_sen}")
```
