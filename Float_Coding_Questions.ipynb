```python
# Question: Calculate the area of a circle with radius 5.5
import math
radius = 5.5
area = math.pi * radius ** 2
print(f"Radius: {radius}")
print(f"Area: {area:.2f}")
```

```python
# Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577

num=input("enter a numbers separated by comma").split(',')
num_list=[]
total=0
for i in num:
    num_list.append(float(i))
for i in num_list:
    total+=i
average=round(total/len(num_list))
print(f"The average of list is {average}")

```

```python
# Question 2: Convert 98.6 Fahrenheit to Celsius (F = C * 9/5 + 32)
F=float(input("enter Fahrenheit"))
C=(F-32)*5/9
F= C * 9/5 + 32
print(f"Fahrenheit to Celsius {C}")
print(f"Celsius to Fahrenheit {F}")
```

```python
# Question 3: Calculate the compound interest on $1000 at 5.5% for 3 years
principal_amount=int(input("enter a principal amount"))
rate_intrest=float(input("enter the rate intrest"))
time=int(input("enter a time"))
final_amount=round(principal_amount*(1+rate_intrest/100)**time)
compound_intrest=final_amount-principal_amount
print(f"The compound intrest is {compound_intrest}")
```

```python
# Question 4: Find the hypotenuse of a right triangle with sides 3.5 and 4.2
import math
a=float(input("enter a value "))
b=float(input("enter b value "))
hypot=round(math.sqrt(a**2+b**2))
print(f"The hypotenuse of a right triangle with sides {hypot}")
```

```python
# Question 5: Calculate the volume of a sphere with radius 7.8
import math
radius=float(input("enter a radius"))
volume=round(4/3*(math.pi*radius**3))
print(f"The volume of sphere is {volume} ")
```

```python
# Question 9: Calculate the simple interest: Principal=2500, Rate=6.5%, Time=2.5 years

principal_amount=int(input("enter a principal amount"))
rate_intrest=float(input("enter the rate intrest"))
time=float(input("enter a time"))
simple_intrest=(principal_amount*rate_intrest*time)/100
print(f"The simple intrest is {simple_intrest}")
```
