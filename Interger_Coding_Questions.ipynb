```python
# Question: Calculate the sum of first 5 even numbers
first_5_even = [2, 4, 6, 8, 10]
total=0
for i in first_5_even:
    total+=i
print(f"First 5 even numbers: {first_5_even}")
print(f"Sum: {total}")
```

```python
# Question 6: Check if 97 is a prime number
number=int(input("enter a number"))
if number<2:
    print(f"{number} is not a prime number")
elif number>=2:
    for i in range(2,number):
        if number%i==0:
            print(f"{number} is not a prime number")
            break
    else:
        print(f"{number} is prime number")
```

```python
# Question 7: Find the factorial of 8
num=int(input("enter a number"))
fact=1
for i in range(1,num+1):
    fact*=i
print(f"The factorial of a number is {fact}")

```

```python
# Question 8: Calculate the average of numbers: 15, 23, 31, 42, 56

numbers_str=input("enter a number separated by comma").split(',')
number_list=[]
total=0
for i in numbers_str:
    number_list.append(int(i))
for num in number_list:
    total=total+num
average=total/len(number_list)
print(f"The average of number list is {average}")
```
