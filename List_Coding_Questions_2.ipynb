```python
# Question 18: Find the sum of all elements in 3D list [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
list_3d=[[[1,2],[3,4]],[[5,6],[7,8]]]
empty=[]
for sublist in list_3d:
    for inner_list in sublist:
        empty.append(sum(inner_list))
        total_sum=sum(empty)
print(total_sum)
```

```python
# Question 19: Extract all even numbers from nested list [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
list=[[1,2,3],[4,5,6],[7,8,9]]
even_nums=[]
for i in list:
    for j in i:
        if j%2==0:
            even_nums.append(j)
print(even_nums)
        
```
