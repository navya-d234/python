```python
# Question 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]
list_nested=[1, [2, 3], [4, [5, 6]], 7]
def count_depth(lst):
    if isinstance(lst,list):
        return 1+max(count_depth(item) for item in lst)
    else:
        return 0
depth=count_depth(list_nested)
print(depth)

```
