```python
# Question 39: Find the most common word in "the quick brown fox jumps over the lazy dog"
str=input("enter string")
words=str.split()
word_count={}
for word in words:
    word_count[word]=word_count.get(word, 0) + 1
most_common_word=max(word_count, key=word_count.get)
print("The most common word is:", most_common_word)
```

```python
# Question 41: Check if string contains balanced parentheses: "((()))"
str=input("enter string")
stack=[]
balanced=True
for char in str:
    if char=='(':
        stack.append(char)
    elif char==')':
        if not stack:
            balanced=False
            break
        stack.pop()
if not stack and balanced:
    print("String has balanced parentheses")
else:
    print("String does not have balanced parentheses")
    
```

```python
# Question 42: Convert "hello world" to Morse code
str="hello world"
morse_code_dict={
    'a':'.-',
    'b':'-...',
    'c':'-.-.',
    'd':'-..',
    'e':'.',
    'f':'..-.',
    'g':'--.',
    'h':'....',
    'i':'..',
    'j':'.---',
    'k':'-.-',
    'l':'.-..',
    'm':'--',
    'n':'-.',
    'o':'---',
    'p':'.--.',
    'q':'--.-',
    'r':'.-.',
    's':'...',
    't':'-',
    'u':'..-',
    'v':'...-',
    'w':'.--',
    'x':'-..-',
    'y':'-.--',
    'z':'--..'
}
morse_code=' '.join(morse_code_dict[char] for char in str if char in morse_code_dict)
print(morse_code)
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"
str1="programming"
str2="grammar"
longest_common=""
for i in range(len(str1)):
    for j in range(i+1, len(str1)+1):
        substring=str1[i:j]
        if substring in str2 and len(substring) > len(longest_common):
            longest_common=substring
print("Longest common substring:", longest_common)
```

```python
# Question 46: Convert "hello world" to Pig Latin
str=input("enter string")
def pig_latin(word):
    first_letter=word[0].lower()
    if first_letter in 'aeiou':
        return word + 'yay'
    else:
        return word[1:] + first_letter + 'ay'
pig_latin_words=[pig_latin(word) for word in str.split()]
print(' '.join(pig_latin_words))
```

```python
# Question 49: Convert "hello world" to ROT13 encoding
str=input("enter string")
def rot13(char):
    if 'a' <= char <= 'z':
        return chr((ord(char) - ord('a') + 13) % 26 + ord('a'))
    elif 'A' <= char <= 'Z':
        return chr((ord(char) - ord('A') + 13) % 26 + ord('A'))
    else:
        return char

rot13_string = ''.join(rot13(c) for c in str)
print(rot13_string)
```

```python
# Question 50: Check if string is a valid credit card number: "4532015112830366"
str=input("enter string")
def luhn_check(card_number):
    total=0
    reverse_digits=card_number[::-1]
    for i, digit in enumerate(reverse_digits):
        n=int(digit)
        if i%2==1:
            n*=2
            if n>9:
                n-=9
        total+=n
    return total%10==0

print(luhn_check(str))
```
